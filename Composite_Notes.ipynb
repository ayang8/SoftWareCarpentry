Some common command options
- data.info()
- !more whatever.csv
- data.T
    - reverses/transposes the columns and rows
- data.describe
    - gives you count, mean, std, min, 25%, 50%, 75%, max
- data.head()
- data.tail()
- data.to_csv(filename.csv)
    - makes .csv file for the wanted table
- data.iloc[]
    - uses numbers within the brackets
    - first numbers refers to rows, numbers after comma refer to columns.
    - use colon (:) if [applies to data.loc as well]
        - you want all rows/columns
        - you want a set of rows/columns by putting in between row or column numbers.
    - when using data.iloc for rectangular tables, does not include "stop number" (see example below)
- data.loc['','']
- data.mean()
    - 0 in the parenthesis gives you means down the rows for all columns
    - 1 in the parenthesis gives you means across the column for all rows
- data.drop('')
    - usually drops a row
    - if you add , axis = 1 after the column you want to drop, you drop a column
- data.idxmax() or data.idxmin()
    - gives you the max and min for each column in terms of the index
- data.sort_values() gives you from smallest to largest
    - .sort_values(ascending = False) gives you from largest to smallest


axis=1 refers to calculations across columns
axis=0 refers to calculation down rows

```python
import pandas as pd
data = pd.read_csv('data/gapminder_gdp_europe.csv', index_col = 'country')
data.loc["Italy":"Poland", 'gdpPercap_1962':'gdpPercap_1972'].max()
```

```python
data.loc["Italy":"Poland", 'gdpPercap_1962':'gdpPercap_1972'].min()
```

```python
subset=data.loc["Italy":"Poland", 'gdpPercap_1962':'gdpPercap_1972']
subset>10000
```

```python
mask=subset>10000
subset[mask]
```

```python
subset[subset>10000].describe()
```

```python
mask_higher = data > data.mean()
wealth_score = mask_higher.aggregate('sum',axis=1)/len(data.columns)
wealth_score

#when summing, trues are labeled 1 and falses labeled 0.
```

```python
data.groupby(wealth_score).mean()
```

```python
#example that iloc does not include "stop number" while loc does
iloc = data.iloc[0:2, 0:2]
print(iloc)
loc = data.loc['Albania':'Belgium', 'gdpPercap_1952':'gdpPercap_1962']
print(loc)
```
